```python
import pandas as pd
from sklearn.datasets import load_digits 
```

```python
digits = load_digits()
```

```python
import matplotlib.pyplot as plt
plt.gray()
for i in range (4):
    plt.matshow(digits.images[i])
```

```python
digits.data[:5]
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
digits.target
```

```python
df['target'] = digits.target
```

```python
df
```

```python
inputs = df.drop(['target'] , axis = 'columns')
```

```python
output = df['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(inputs , output , test_size = 0.2 , random_state = 48)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_model = RandomForestClassifier()
```

```python
random_model.fit(X_train , y_train)
```

```python
random_model.score(X_test , y_test)
```

```python
#confusion matrix  
```

```python
y_predicted = random_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cn = confusion_matrix(y_predicted , y_test)
```

```python
cn
```
